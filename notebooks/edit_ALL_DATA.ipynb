# 목차 데이터는 페이지 내용에서 삭제

```python
# 실행 완료

for file_name, page_info_dict in index_pages.items():
    if file_name not in ALL_DATA.keys():
        continue

    index_page_label = page_info_dict["index_page_label"]

    count = len(index_page_label)

    if count != 0:
        for dict_ in ALL_DATA[file_name]:
            metadata = dict_["metadata"]
            if metadata["page"] in index_page_label:
                dict_["content"] = ""
                count -= 1

            if count == 0:
                break

    start_page_label = page_info_dict["start_page_label"]

    for dict_ in ALL_DATA[file_name]:
        if dict_["metadata"]["page"] >= start_page_label:
            dict_["metadata"]["page"] -= start_page_label - 1
        else:
            dict_["metadata"]["page"] = None
```

```python
import os
new_data_path = all_data_path.replace(".json", "_new.json")
with open(new_data_path, "w", encoding="utf-8") as f:
    json.dump(ALL_DATA, f, indent=2, ensure_ascii=False)
```
